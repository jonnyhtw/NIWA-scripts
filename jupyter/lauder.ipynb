```python
fname = '/home/'+os.environ['USER']+'/scripts/foo'

ozonestash = iris.AttributeConstraint(STASH='m01s34i001') 
brostash = iris.AttributeConstraint(STASH='m01s34i994') 
no2stash = iris.AttributeConstraint(STASH='m01s34i996') 

ozone = iris.load(fname,ozonestash)
bro = iris.load(fname,brostash)
no2 = iris.load(fname,no2stash)
```

```python
lon = 169-180
lat = -45

markersize = 10

cmap1 = plt.get_cmap('viridis_r')
cmap2 = plt.get_cmap('inferno_r')
cmap3 = plt.get_cmap('magma_r')

plt.figure(figsize=(12, 5))
proj = ccrs.PlateCarree(central_longitude=180.0)
plt.subplot(111, projection=proj)
iplt.pcolormesh(ozone[0][-1][0][:][:],cmap=cmap1)
plt.plot(lon, lat, 'bX', markersize=markersize)
plt.gca().coastlines()
#plt.colorbar()
plt.title('O$_3$ at 20m')
iplt.show()

plt.figure(figsize=(12, 5))
proj = ccrs.PlateCarree(central_longitude=180.0)
plt.subplot(111, projection=proj)
iplt.pcolormesh(bro[0][-1][0][:][:],cmap=cmap2)
plt.plot(lon, lat, 'bX', markersize=markersize)
plt.gca().coastlines()
#plt.colorbar()
plt.title('BrO at 20m')
iplt.show()

plt.figure(figsize=(12, 5))
proj = ccrs.PlateCarree(central_longitude=180.0)
plt.subplot(111, projection=proj)
iplt.pcolormesh(no2[0][-1][0][:][:],cmap=cmap3)
plt.plot(lon, lat, 'bX', markersize=markersize)
plt.gca().coastlines()
#plt.colorbar()
plt.title('NO$_2$ at 20m')
iplt.show()
```
