```python
import iris
import matplotlib.pyplot as plt
import glob
```

```python
my_cell_method = (iris.coords.CellMethod('mean', 'time', intervals='1 hour'),)
```

```python
cell_constr=iris.Constraint(cube_func=lambda cube: cube.cell_methods == my_cell_method)
```

```python
cd '/home/williamsjh/cylc-run/u-bj476/share/data/History_Data/'
```

```python
stash_constraint = iris.AttributeConstraint(STASH='m01s00i024')
```

```python
cubes = iris.load(glob.glob('./*pm*195[0-7]*'), stash_constraint)
```

```python
bj476 = cubes.extract(cell_constr)
```

```python
bj476[0].coord('longitude').guess_bounds()
bj476[0].coord('latitude').guess_bounds()
```

```python
 grid_areas = iris.analysis.cartography.area_weights(bj476[0])
```

```python
bj476mean = bj476[0].collapsed(['longitude', 'latitude'], iris.analysis.MEAN, weights=grid_areas)
```
